```python
import os
import random
from PIL import Image
import matplotlib.pyplot as plt

# Define the path to the dataset
data_dir = r'C:\Users\rohin\Documents\Fall-2024\DeepLearning\DL-recyclable-Waste-Recognition\dataset'  # Adjusted to point to the dataset folder
print(data_dir)
categories = ['cardboard', 'glass', 'metal', 'paper', 'plastic', 'trash']

# Function to display sample images from each category
def display_samples():
    fig, axs = plt.subplots(1, len(categories), figsize=(15, 5))
    fig.suptitle('Sample Images from Each Category')
    
    for i, category in enumerate(categories):
        category_path = os.path.join(data_dir, category)
        if not os.path.exists(category_path):
            print(f"Warning: Category folder {category_path} does not exist.")
            continue

        # Randomly select an image from the category
        sample_image = random.choice(os.listdir(category_path))
        image_path = os.path.join(category_path, sample_image)
        
        img = Image.open(image_path)
        axs[i].imshow(img)
        axs[i].set_title(category)
        axs[i].axis('off')
    
    plt.show()

# Function to display dataset distribution
def display_distribution():
    counts = {category: 0 for category in categories}
    
    for category in categories:
        category_path = os.path.join(data_dir, category)
        
        if not os.path.exists(category_path):
            print(f"Warning: Category folder {category_path} does not exist.")
            continue
        
        counts[category] = len(os.listdir(category_path))
    
    plt.figure(figsize=(10, 6))
    plt.bar(counts.keys(), counts.values(), color='skyblue')
    plt.title('Dataset Distribution by Category')
    plt.xlabel('Category')
    plt.ylabel('Number of Images')
    plt.show()

# Function to display a random image with bounding box (if applicable)
def display_random_sample_with_box():
    random_category = random.choice(categories)
    category_path = os.path.join(data_dir, random_category)
    
    if not os.path.exists(category_path):
        print(f"Warning: Category folder {category_path} does not exist.")
        return
    
    # Randomly select an image from the category
    image_file = random.choice(os.listdir(category_path))
    image_path = os.path.join(category_path, image_file)
    
    img = Image.open(image_path)
    plt.imshow(img)
    plt.title(f"Sample Image: {image_file}")
    plt.axis('off')
    plt.show()

# Run visualizations
display_samples()  # Show sample images from the dataset
display_distribution()  # Show the dataset distribution
display_random_sample_with_box()  # Show a random sample with YOLO bounding box (if applicable)
```
